```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:\IPCV\images\eagle_img.webp',0)
```

```python
f = np.fft.fft2(img)
```

```python
fshift = np.fft.fftshift(f)
```

```python
row, cols = img.shape
```

```python
crow,ccol = row//2, cols//2
```

```python
mask_lp = np.zeros((row, cols), np.uint8)
```

```python
mask_lp[crow-30:crow+30,ccol-30:ccol+30] = 1
```

```python
lp = np.fft.ifft2(np.fft.ifftshift(fshift*mask_lp))
```

```python
cv2.imwrite('lpf_egale.png',np.abs(lp))
```

```python
mask_hp = 1 - mask_lp
```

```python
hp = np.fft.ifft2(np.fft.ifftshift(fshift*mask_hp))
```

```python
cv2.imwrite('hpf_egale.png',np.abs(hp))
```
