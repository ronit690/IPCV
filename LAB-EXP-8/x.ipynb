```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'D:\IPCV\LAB-EXP-8\images\2.jpg')
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
h,s,v = cv2.split(hsv)
```

```python
print(hsv)
```

```python
lower_h = 35
upper_h = 85
```

```python
mask = cv2.inRange(h,lower_h,upper_h)
```

```python
result = cv2.bitwise_and(img,img,mask=mask)
```

```python
print(mask)
```

```python
cv2.imwrite('mask.jpg',result)
```

# Add Saturation and Value filtering

```python
lower_green = np.array([35, 50, 50])   # H, S, V (min)
upper_green = np.array([85, 255, 255]) # H, S, V (max)
```

```python
mask = cv2.inRange(hsv, lower_green, upper_green)
```

```python
result = cv2.bitwise_and(img, img, mask=mask)
```

```python
# Save original image
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\original.jpg', img)

# Save HSV image (note: colors will look different if opened normally)
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\hsv.jpg', hsv)

# Save individual channels
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\hue.jpg', h)
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\saturation.jpg', s)
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\value.jpg', v)

# Save final filtered result
cv2.imwrite(r'D:\IPCV\LAB-EXP-8\filtered_green.jpg', result)
```
