```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# ─────────────────────────────────────────
# Step 1: Read Original & Create S&P Noise
# ─────────────────────────────────────────
img = cv2.imread(r'D:\IPCV\LAB-EXP-5\images\1.jpeg', 0)

sp = img.copy().astype(np.float32)
prob = 0.02
rand = np.random.rand(*img.shape)
sp[rand < prob] = 0
sp[rand > 1 - prob] = 255
sp = sp.astype(np.uint8)

# ─────────────────────────────────────────
# Step 2: Histogram Equalization
# ─────────────────────────────────────────
hist_eq = cv2.equalizeHist(sp)

# ─────────────────────────────────────────
# Step 3: Sharpening Filter
# ─────────────────────────────────────────
kernel = np.array([[0, -1,  0],
                   [-1,  5, -1],
                   [0, -1,  0]])
sharpened = cv2.filter2D(sp, -1, kernel)

# ─────────────────────────────────────────
# Step 4: Contrast Stretching
# ─────────────────────────────────────────
min_val = np.min(sp)
max_val = np.max(sp)
contrast_stretched = ((sp - min_val) / (max_val - min_val) * 255).astype(np.uint8)

# ─────────────────────────────────────────
# Step 5: Plot All
# ─────────────────────────────────────────
plt.figure(figsize=(15, 5))

plt.subplot(1, 5, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(sp, cmap='gray')
plt.title('Salt & Pepper Noise')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(hist_eq, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpening Filter')
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretching')
plt.axis('off')

plt.tight_layout()
plt.savefig(r'D:\IPCV\LAB-EXP-5\resultant_img\enhanced_sp.jpeg')
plt.show()

# ─────────────────────────────────────────
# Step 6: Save All Enhanced Images
# ─────────────────────────────────────────
os.makedirs(r'D:\IPCV\LAB-EXP-5\resultant_img', exist_ok=True)

cv2.imwrite(r'D:\IPCV\LAB-EXP-5\resultant_img\6_hist_equalized.jpeg',     hist_eq)
cv2.imwrite(r'D:\IPCV\LAB-EXP-5\resultant_img\7_sharpened.jpeg',          sharpened)
cv2.imwrite(r'D:\IPCV\LAB-EXP-5\resultant_img\8_contrast_stretched.jpeg', contrast_stretched)

print("All enhanced images saved!")
```
