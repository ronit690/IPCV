```python
import cv2
```

```python
img = cv2.imread(r'D:\IPCV\LAB-EXP-9\images\1.jpg', 0)
```

```python
_, binary = cv2.threshold(img,127,225,cv2.THRESH_BINARY)
```

```python
adaptive = cv2.adaptiveThreshold(
    img,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)
```

```python
cv2.imwrite(r'D:\IPCV\LAB-EXP-9\images\binary.jpg', binary)
cv2.imwrite(r'D:\IPCV\LAB-EXP-9\images\adaptive.jpg', adaptive)
```
